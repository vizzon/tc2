## Calcular los parametros Z de la siguiente red

<img src="./assets/parte_1.png" align="center"/> 

En primer lugar recordamos como se compone la matrix Z de una red:

$\begin{equation}
\begin{vmatrix}
V_1\\
V_2\\
\end{vmatrix}
= 
\begin{vmatrix}
Z_{11} & Z_{12}\\
Z_{21} & Z_{22}\\
\end{vmatrix}
\begin{vmatrix}
I_1\\
I_2\\
\end{vmatrix}
\end{equation}$

Operando y dejando a circuito abierto la salida o la entrada segun corresponda llegamos a:

$Z_{11} = \frac{V_1}{I_1} \land I_2 = 0 $

$Z_{12} = \frac{V_1}{I_2} \land I_1 = 0 $

$Z_{21} = \frac{V_2}{I_1} \land I_2 = 0 $

$Z_{22} = \frac{V_2}{I_2} \land I_2 = 0 $

Nombramos a la corriente que entra por el primario del transformador como $I_{L1}$ y a la que sale por el secundario del transformador $I_{L2}$. 
Como la relacion del transformador es 1:1 (con acoplamiento ideal) y la disposicion de los bornes homologos hace que la corriente que entra por el borne superior del primario salga por el borne superior del secundario podemos decir que:
$I_{L1} = I_{L2} = I_L$

Ahora planteamos las ecuaciones del circuito:

$V_1 = -V_2$

$I_1 = I_{R_1} + I_{R_2} + I_L = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + I_L$

$I_2 = I_{R_3} + I_{R_2} + I_L = \frac{V_2}{R_3} + \frac{V_2 - V_1}{R_2} + I_L = \frac{V_2}{R_3} - \frac{V_1 - V_2}{R_2} + I_L$

### Calculo de $Z_{11}$

$Z_{11} = \frac{V_1}{I_1} \land I_2 = 0 $

$I_2 = \frac{V_2}{R_3} - \frac{V_1 - V_2}{R_2} + I_L = 0 \implies I_L = \frac{V_1 - V_2}{R_2} - \frac{V_2}{R_3}$

$I_1 = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + I_L = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + \frac{V_1 - V_2}{R_2} - \frac{V_2}{R_3} = \frac{V_1}{R_1} + \frac{V_1 + V_1}{R_2} + \frac{V_1 + V_1}{R_2} + \frac{V_1}{R_3} = \frac{V_1}{R_1} + \frac{4V_1}{R_2} + \frac{V_1}{R_3} = V_1 \left(\frac{1}{R_1} + \frac{4}{R_2} + \frac{1}{R_3}\right)$

$I_1 = V_1 \left(\frac{1}{R_1} + \frac{4}{R_2} + \frac{1}{R_3}\right)$

Reemplazando $R_1 = 1 , R_2 = 2 , R_3 = 3 $

$I_1 = V_1 \left(\frac{1}{1} + \frac{4}{2} + \frac{1}{3}\right) = V_1 \left(3 + \frac{1}{3}\right) = V_1 \left(\frac{10}{3}\right) \implies \frac{V_1}{I_1} = \frac{3}{10} = 0.3 = Z_{11}$

### Calculo de $Z_{12}$

$Z_{12} = \frac{V_1}{I_2} \land I_1 = 0 $

$I_1 = 0 = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + I_L \implies I_L = -\frac{V_1}{R_1} - \frac{V_1 - V_2}{R_2}$

$I_2 = \frac{V_2}{R_3} - \frac{V_1 - V_2}{R_2} -\frac{V_1}{R_1} - \frac{V_1 - V_2}{R_2} = -\frac{V_1}{R_3} - \frac{2V_1}{R_2} -\frac{V_1}{R_1} - \frac{2V_1}{R_2} = -\frac{V_1}{R_3} - \frac{4V_1}{R_2} -\frac{V_1}{R_1}$

Agrupando y reemplazando los valores de los componentes

$I_2 = -V_1 \left(\frac{1}{3} + \frac{4}{2} + 1\right) = -V_1 \left(\frac{10}{3}\right) \implies \frac{V_1}{I_1} = -\frac{3}{10} = Z_{12}$ 

### Calculo de $Z_{21}$

$Z_{21} = \frac{V_2}{I_1} \land I_2 = 0 $

$I_2 = \frac{V_2}{R_3} - \frac{V_1 - V_2}{R_2} + I_L = 0 \implies I_L = \frac{V_1 - V_2}{R_2} - \frac{V_2}{R_3}$

$I_1 = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + I_L = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + \frac{V_1 - V_2}{R_2} - \frac{V_2}{R_3} = -\frac{V_2}{R_1} - \frac{2V_2}{R_2} - \frac{2 V_2}{R_2} - \frac{V_2}{R_3} = -V_2 \left(\frac{1}{R_1} - \frac{4}{R_2} - \frac{1}{R_3}\right)$

Reemplazando los valores de los componentes

$I_1 = -V_2 \left(1 - \frac{4}{2} - \frac{1}{3}\right) = -V_1 \left(\frac{10}{3}\right) \implies \frac{V_2}{I_1} = -\frac{3}{10} = -0.3 = Z_{21}$

### Calculo de $Z_{22}$

$Z_{22} = \frac{V_2}{I_2} \land I_1 = 0 $

$I_1 = 0 = \frac{V_1}{R_1} + \frac{V_1 - V_2}{R_2} + I_L \implies I_L = - \frac{V_1}{R_1} - \frac{V_1 - V_2}{R_2}$

$I_2 = \frac{V_2}{R_3} - \frac{V_1 - V_2}{R_2} - \frac{V_1}{R_1} - \frac{V_1 - V_2}{R_2} = \frac{V_2}{R_3} + \frac{2V_2}{R_2} + \frac{V_2}{R_1} + \frac{2 V_2}{R_2} = V_2 \left(\frac{1}{R_3} + \frac{4}{R_2} + \frac{1}{R_1}\right) $

Reemplazando los valores de los componentes

$I_2 = V_2 \left(\frac{1}{3} + \frac{4}{2} + 1\right) = V_2\frac{10}{3} \implies \frac{V_2}{I_2} = \frac{3}{10} = 0.3 = Z_{22}$

## Calculo de la transferencia $\frac{V_2}{V_1}$ por MAI

$V^{ij}_{mn} = \frac{V_{ij}}{V_{mn}} = sgn(i-j)sgn(m-n) \frac{\underline{Y}^{ij}_{mn}}{\underline{Y}^{mn}_{mn}}$

$\underline{Y}^{ij}_{mn} = (-1)^{m+n+i+j}\Delta Y^{ij}_{mn}$

$\Delta Y^{ij}_{mn}$ es el determinante de la MAI con las columnas i,j suprimidas y las filas m,n suprimidas.

$V^{34}_{14} = \frac{V_{34}}{V_{14}} = (-1)(-1)\frac{\underline{Y}^{34}_{14}}{\underline{Y}^{14}_{14}}$

$\underline{Y}^{14}_{14} = (-1)^{1+4+1+4} \times\left( \frac{4s^2+8}{3s}\frac{s+2}{s} - \frac{4}{s^2}\right) = 
\frac{4s^3 + 8s^2 + 8s + 16}{3s^2} - \frac{4}{s^2} = \frac{4s^3 + 8s^2 + 8s + 16 - 12}{3s^2} = \frac{4s^3 + 8s^2 + 8s + 4}{3s^2}$

$\underline{Y}^{14}_{34} = (-1)^{1+4+3+4} \times \left(\frac{-2}{3s}\frac{-2}{s}\right) = \frac{4}{3s^3}$

$\frac{V_2}{V_1} = \frac{V_{34}}{V_{14}} = \frac{4}{3s^3} \times \frac{3s^2}{4s^3 + 8s^2 + 8s + 4} =
 \frac{4}{3s^3} \times \frac{3s^2}{4}\frac{1}{s^3 + 2s^2 + 2s + 1} = \frac{1}{s^3 + 2s^2 + 2s + 1} = \frac{V_2}{V_1}$

Se observa que lo obtenido por interconeccion de cuadripolos y lo obtenido por el metodo de MAI es consistente.
